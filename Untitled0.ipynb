```python
# hydraulic_fracture_simulation.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1) Project Parameters
# -----------------------------
GRID_SIZE      = 80        # 80x80 reservoir grid
TIME_STEPS     = 250       # simulation time steps
DIFFUSIVITY    = 0.2      # pressure spread rate
INJ_PRESSURE   = 4.5       # injection pressure at well
FRACTURE_CRIT  = 1.0       # threshold for fracture
LEAKOFF        = 0.002     # leakoff factor

# -----------------------------
# 2) Initialize Data Structures
# -----------------------------
pressure_grid  = np.zeros((GRID_SIZE, GRID_SIZE))
fracture_grid  = np.zeros((GRID_SIZE, GRID_SIZE))

well_i = GRID_SIZE // 2
well_j = GRID_SIZE // 2

fracture_counts = []
avg_pressures   = []

# -----------------------------
# 3) Simulation Loop
# -----------------------------
for step in range(TIME_STEPS):
    next_pressure = pressure_grid.copy()

    # Continuous injection
    pressure_grid[well_i, well_j] = INJ_PRESSURE

    # Diffusion (simplified finite-difference)
    for i in range(1, GRID_SIZE-1):
        for j in range(1, GRID_SIZE-1):
            laplacian = (
                pressure_grid[i+1, j] +
                pressure_grid[i-1, j] +
                pressure_grid[i, j+1] +
                pressure_grid[i, j-1] -
                4 * pressure_grid[i, j]
            )
            next_pressure[i,j] = pressure_grid[i,j] + DIFFUSIVITY*laplacian - LEAKOFF*pressure_grid[i,j]

    pressure_grid = next_pressure

    # Fracture condition
    fracture_grid[pressure_grid >= FRACTURE_CRIT] = 1

    # Store stats
    fracture_counts.append(np.sum(fracture_grid))
    avg_pressures.append(np.mean(pressure_grid))

# -----------------------------
# 4) Save Results to CSV
# -----------------------------
df = pd.DataFrame({
    "Time Step": range(TIME_STEPS),
    "Average Pressure": avg_pressures,
    "Fracture Count": fracture_counts
})
df.to_csv("fracking_sim_results.csv", index=False)

# -----------------------------
# 5) Visualization
# -----------------------------
plt.figure()
plt.imshow(fracture_grid, cmap="hot")
plt.title("Fracture Propagation Map")
plt.colorbar(label="Fracture Flag (0/1)")
plt.xlabel("Grid X")
plt.ylabel("Grid Y")
plt.savefig("fracture_map.png")
plt.show()

plt.figure()
plt.plot(fracture_counts, label="Fractured Area")
plt.plot(avg_pressures, label="Mean Pressure")
plt.title("Fracture Growth & Pressure Over Time")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.legend()
plt.savefig("growth_curve.png")
plt.show()

print("Simulation Completed.")
print("Final Fractured Cells:", fracture_counts[-1])
```
